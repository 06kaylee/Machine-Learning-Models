Logistic Regression

Import libraries

```
import numpy as np
import pandas as pd
```

Load in the dataset

```
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values   # X stores the matrix of features which contains all the rows of every column except for the last column
y = dataset.iloc[:, -1].values    # y stores the dependent variable (what you're predicting). It contains all the rows of the last column
```

Split dataset into training and test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

Feature scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()   # StandardScaler calculates the mean and standard deviation for a feature. Then scales the feature: (feature - mean(feature)) / std(feature)
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

Train the model

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

Predict new results from test set

```
y_pred = classifier.predict(X_test)
```

Evaluate model

```
from sklearn.metrics import confusion_matrix, accuracy_score
c_matrix = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print(f"Confusion matrix: {c_matrix}")
print(f"Accuracy score: {acc}")
```
